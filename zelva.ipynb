```python
from ColabTurtle.Turtle import *
initializeTurtle(initial_speed=13,initial_window_size=(2400,2400))

color('orange')
width(10)


for _ in range(18):
    forward(140)
    left(20)
for _ in range(3):
    forward(140)
    left(20)
width(15)
for _ in range(8):
    left(-45)
    forward(100)
for _ in range(4):
    forward(140)
    left(20)
for _ in range(8):
    left(-45)
    forward(100)
penup()
goto(650,1000)
pendown()
for _ in range(18):
    forward(20)
    left(20)
penup()
goto(900,1000)
pendown()
for _ in range(18):
    forward(20)
    left(20)
penup()
goto(650,1200)
pendown()
for _ in range(6):
    left(37)
    forward(50)
penup()
goto(800,1200)
pendown()
for _ in range(1):
    backward(50)
for _ in range(5):
    left(35)
    backward(50)
penup()
goto(670,750)
pendown()
color('red')
for _ in range(1):
    right(-160)
    forward(300)
color('white')
for _ in range(18):
    forward(20)
    left(20)
color('red')
for _ in range(1):
    right(130)
    forward(300)


color('yellow')
width(25)
penup()
goto(1800,200)
pendown()
for _ in range(9):
    forward(120)
    left(80)


color('yellow')
width(25)
penup()
goto(1400,500)
pendown()
for _ in range(9):
    forward(120)
    left(80)

color('yellow')
width(25)
penup()
goto(300,300)
pendown()
for _ in range(9):
    forward(120)
    left(80)

color('white')
width(3)
hideturtle()

def draw_snowflake_branch(size):

    for _ in range(3):
        forward(size/3)
        right(30)
        forward(size/6)
        backward(size/6)
        left(60)
        forward(size/6)
        backward(size/6)
        right(30)
    backward(size)

def draw_snowflake(x, y, size, num_branches=6):
    penup()
    goto(x, y)
    pendown()
    for _ in range(num_branches):
        draw_snowflake_branch(size)
        left(360 / num_branches)

draw_snowflake(200, 1800, 50)
draw_snowflake(1100, 200, 60)
draw_snowflake(950, 1100, 45)
draw_snowflake(2200, 850, 55)
draw_snowflake(800, 1000, 70)
draw_snowflake(1200, 1500, 50)
draw_snowflake(900, 800, 40)
draw_snowflake(100, 1150, 65)
draw_snowflake(980, 980, 55)
draw_snowflake(1020, 1020, 48)
draw_snowflake(1500, 500, 60)
draw_snowflake(1800, 1200, 42)
draw_snowflake(500, 500, 75)
draw_snowflake(2000, 2000, 60)
draw_snowflake(300, 100, 55)
draw_snowflake(1600, 100, 65)
draw_snowflake(1000, 2200, 70)
```
